```python


def draw_chessboard_with_border(ax, tiles, title):
    # Dimensione della scacchiera e dello spessore del bordo
    n = 5
    border_width = 0.05  # Spessore del bordo

    # Disegna la scacchiera
    for x in range(n):
        for y in range(n):
            if (x + y) % 2 == 0:
                ax.add_patch(patches.Rectangle((x, y), 1, 1, color='lightgray'))
            else:
                ax.add_patch(patches.Rectangle((x, y), 1, 1, color='darkgray'))

    # Aggiungi le tessere alla scacchiera
    for tile in tiles:
        ax.add_patch(patches.Rectangle(tile['pos'], tile['width'], tile['height'], 
                                       edgecolor='black', facecolor=tile['color'], linewidth=2))

    # Aggiungi il bordo attorno alla scacchiera
    ax.add_patch(patches.Rectangle((-border_width, -border_width), n + 2*border_width, n + 2*border_width, 
                                   fill=False, edgecolor='black', linewidth=2))

    # Imposta i limiti degli assi considerando lo spessore del bordo
    ax.set_xlim(-border_width, n + border_width)
    ax.set_ylim(-border_width, n + border_width)
    ax.set_aspect('equal')

    # Nascondi gli assi
    ax.axis('off')

    # Titolo
    ax.set_title(title, fontsize=30, pad=15, color='black', fontweight='bold')

# Dati per le 4 scacchiere
scacchiera1 = [{'pos': (3, 1), 'width': 3, 'height': 1, 'color': 'skyblue'},
               {'pos': (1, 3), 'width': 3, 'height': 1, 'color': 'lightgreen'}]

scacchiera2 = [{'pos': (1, 1), 'width': 3, 'height': 1, 'color': 'skyblue'},
               {'pos': (3, 3), 'width': 1, 'height': 3, 'color': 'lightgreen'}]

scacchiera3 = [{'pos': (2, 2), 'width': 3, 'height': 1, 'color': 'skyblue'},
               {'pos': (3, 1), 'width': 1, 'height': 3, 'color': 'lightgreen'}]

scacchiera4 = [{'pos': (1, 0), 'width': 3, 'height': 1, 'color': 'skyblue'},
               {'pos': (3, 2), 'width': 1, 'height': 3, 'color': 'lightgreen'}]

# Crea una figura per contenere tutte le scacchiere
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Disegna le 4 scacchiere con il bordo
draw_chessboard_with_border(axs[0][0], scacchiera1, "Tessera eccede i limiti \norizzontalmente")
draw_chessboard_with_border(axs[0][1], scacchiera2, "Tessera eccede i limiti \nverticalmente")
draw_chessboard_with_border(axs[1][0], scacchiera3, "Sovrapposizione delle tessere")
draw_chessboard_with_border(axs[1][1], scacchiera4, "Posizionamento corretto")

plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.4)

#plt.tight_layout()

plt.savefig("scacchiere_vincoli_base.png",bbox_inches='tight')
plt.show()
```
